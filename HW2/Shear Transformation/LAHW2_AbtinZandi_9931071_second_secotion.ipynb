# Shadow Of Image

In this method, we make its shadow image based on non white colors

```python
def shadow_of_image(img):
    temp_copy_of_image = np.copy(img)
    for index in range (0,len(temp_copy_of_image)):
        for secondIndex in range (0,len(temp_copy_of_image[index])):
            for colors in range (0,len(temp_copy_of_image[index,secondIndex])):
                #we want to remove colors to gray shadow
                #249 ... 255 are rgb of white and 198 is rgb of shadow
                if temp_copy_of_image[index,secondIndex,colors] < 249:
                    temp_copy_of_image[index,secondIndex,colors] = 198

    plt.imshow(temp_copy_of_image)
    plt.title("Shadow Image")
    plt.show()
    type(temp_copy_of_image)
    return temp_copy_of_image
```

```python
second_picture_shadow = shadow_of_image(second_img)
```

# Transform and Combination (image and image shadow)

In this function, we first calculate the diagonal displacement of the shadow relative to the image and add the first image to it

```python
def Transform_and_combination_of_image(img,picture_shadow):
    Transform_image = np.arange(12*len(img)*len(img[0])).reshape(2*len(img),2*len(img[0]) ,3)
    Final_Image = np.arange(12*len(img)*len(img[0])).reshape(2*len(img),2*len(img[0]) ,3)
    #transform and combine
    len_img = len(img)
    for colors in range (0,3):
        for index in range (0,len_img):
            for secondIndex in range (0,len(img[index])):
                if img[index,secondIndex,colors] < 249:
                        Transform_image[index,secondIndex + index,colors] = picture_shadow[index,secondIndex,colors]
                        Final_Image[index,secondIndex + len_img  ,colors] = img[index,secondIndex,colors]
                        Final_Image[index,secondIndex + index,colors] = picture_shadow[index,secondIndex,colors]
                        
                else : 
                    Transform_image[index,index,colors] = 255
                    Final_Image[index,index,colors] = 255
    plt.subplot(2,1,1)
    plt.imshow(Transform_image)
    plt.title("Transform Image")
    plt.show()
    type(Transform_image)
    
    plt.subplot(2,1,2)
    plt.imshow(Final_Image)
    plt.title("Transformed and Combined Image")
    plt.show()
    type(Transform_image)
```

```python
Transform_and_combination_of_image(second_img,second_picture_shadow)
```
