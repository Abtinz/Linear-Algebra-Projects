Sperating Colors and plotting them

```python
def green_rate_plot(img,rgb_rate):
    green_rate = np.arange(0,256)*0
    for i in range(0,img.shape[0]):
        for j in range(0,img.shape[1]):
            green_rate[img[i,j,1]] += 1
    plt.bar(rgb_rate,green_rate)
    plt.title("green Rate bar plot")
    plt.xlabel("RGB rate")
    plt.ylabel("green Count per pixel")      
```

```python
green_rate_plot(img,rgb_rate)
```

```python
def red_rate_plot(img,rgb_rate):
    red_rate = np.arange(0,256)*0
    for i in range(0,img.shape[0]):
        for j in range(0,img.shape[1]):
            red_rate[img[i,j,0]] += 1
    plt.bar(rgb_rate,red_rate)
    plt.title("red Rate bar plot")
    plt.xlabel("RGB rate")
    plt.ylabel("red Count per pixel")
```

```python
red_rate_plot(img,rgb_rate)
```

```python
def blue_rate_plot(img,rgb_rate):
    blue_rate = np.arange(0,256)*0
    for i in range(0,img.shape[0]):
        for j in range(0,img.shape[1]):
            blue_rate[img[i,j,2]] += 1
    plt.bar(rgb_rate,blue_rate)
    plt.title("blue Rate bar plot")
    plt.xlabel("RGB rate")
    plt.ylabel("blue Count per pixel")  
```

```python
blue_rate_plot(img,rgb_rate)
```
